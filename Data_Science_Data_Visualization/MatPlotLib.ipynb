```python
#Pyplot Tutorial
import matplotlib.pyplot as plt
```

```python
plt.plot([1,4,5,2,6]) #if we pass only one list then it becomes a y axis and x axis is index of array
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title('myplot')
plt.show()
```

```python
#print with grid
#if we pass two lists then first one is x-axis and second one is y-axis
#let see an example
plt.plot([1,2,3,4,5,6],[1,4,9,16,25,36])
plt.xlabel("squares")
plt.ylabel("Numbers")
plt.title("Numbers and corresponding squares")
plt.grid()
plt.show()
```

```python
#another plot instead of line use dot of red color
plt.plot([1,2,3,4,5,6,7,8],[1,4,9,16,25,36,49,64],'ro')
plt.grid()
plt.show()
```

```python
#multile plots in single graph
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0., 7., 0.2)

plt.plot(t,t**2,'b--',label = '^2')
plt.plot(t,t**2.2,'rs',label = '^2.2')
plt.plot(t,t**2.5,'g^',label = '^2.5')
plt.xlabel('Numbers')
plt.ylabel('powers')
plt.grid()
plt.legend()
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y,linewidth = 5.0)
plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [1,4,9,16,25]
x2 = [1,2,3,4,5]
y2 = [2,4,6,8,10]

lines = plt.plot(x1,y1,x2,y2)
#use keywords args
plt.setp(lines[0],color = 'r',linewidth = 2.0)
#matlab style
plt.setp(lines[1],'color','g','linewidth',2.0)
plt.grid()
plt.show()
```

```python
#two plots in a figure
t1 = np.arange(0.0,5.0,0.2)
t2 = np.arange(0.0,10.0,0.4)

plt.figure(1)

plt.subplot(221)
plt.grid()
plt.plot(t1,t2,'b-')

plt.subplot(222)
plt.plot(t2,t1,'r--')



plt.show()
```
