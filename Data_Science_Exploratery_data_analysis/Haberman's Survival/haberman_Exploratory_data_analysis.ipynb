* We have 4 features:

 * Year of treatment(operation) (We have data of 12 years from 1958 to 1969.)
 * positive_nodes(Positive Lymph node- related to breast cancer)
 * survrival_status_after_5_years (survival status after 5 years.)
   * 1- indicate survived for 5 or more years
   * 2- indicate died before 5 years
 

# 2.Univeriate DataAnalysis

## (2.1) PDF(probability Density Function) 

```python
print(cancer_df.columns)
```

```python
# let see survival according to the Age of operation 
sns.set_style("darkgrid")
sns.FacetGrid(cancer_df,hue="survival_status_after_5_years",size=5) \
    .map(sns.distplot,"age") \
    .add_legend()
    
plt.show()
```

```python
#let see survival according to the positive lymph nodes
sns.set_style("whitegrid")
sns.FacetGrid(cancer_df,hue="survival_status_after_5_years",size=6) \
    .map(sns.distplot,"positive_nodes") \
    .add_legend()
plt.show()
```

```python
#let see survival according to the year of operation
sns.set_style("whitegrid")
sns.FacetGrid(cancer_df,hue="survival_status_after_5_years",size=6) \
    .map(sns.distplot,"year_of_treatment") \
    .add_legend()
plt.show()
```
