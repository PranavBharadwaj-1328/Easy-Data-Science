```python
%%time
fig, ax = plt.subplots(1,2,figsize=(30,7))
wordcloud = WordCloud(background_color='black',width=1500,height=1500).generate(' '.join(data['Name']))
wordcloud_sub = WordCloud(background_color='white',width=1500,height=1500).generate(' '.join(data['Subtitle'].dropna().astype(str)))
ax[0].imshow(wordcloud)
ax[0].axis('off')
ax[0].set_title('WordCloud(Name)')
ax[1].imshow(wordcloud_sub)
ax[1].axis('off')
ax[1].set_title('WordCloud(Subtitle)')
plt.show()
```
