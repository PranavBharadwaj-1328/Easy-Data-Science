```python
%%time
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO
```

```python
%%time
fig, ax = plt.subplots(5,5, figsize = (12,12))

for i in range(25):
    r = requests.get(data['Icon URL'][i])
    im = Image.open(BytesIO(r.content))
    ax[i//5][i%5].imshow(im)
    ax[i//5][i%5].axis('off')
plt.show()
```

```python
aur  = data['Average User Rating'].value_counts().sort_index()
p = figure(x_range = list(map(str,aur.index.values)),plot_height = 250, title = "Average User Rating",toolbar_location=None,tools="")
p.vbar(x=list(map(str,aur.index.values)),top=aur.values,width=0.9,color = Spectral9)
p.xgrid.grid_line_color = None
p.y_range.start = 0
show(p)
```
