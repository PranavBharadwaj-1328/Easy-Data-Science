```python
%%time
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO
```

```python
%%time
fig, ax = plt.subplots(5,5, figsize = (12,12))

for i in range(25):
    r = requests.get(data['Icon URL'][i])
    im = Image.open(BytesIO(r.content))
    ax[i//5][i%5].imshow(im)
    ax[i//5][i%5].axis('off')
plt.show()
```

# Price and Rating (Seaborn : regplot,Implot )

I want to see the relationship between price and ratings

```python
fig,ax = plt.subplots(1,1, figsize=(12,7),dpi = 72)
sns.regplot(data = data,x = "Price" , y= "Average User Rating",ax = ax)
plt.show()
```

## Observation :

You can see there are mostly apps with rating 4 and have very less price

```python
price_list = sorted(data['Price'].dropna().unique())
rating_stats = pd.DataFrame(columns=['mean','std','count'],index=price_list)
for price in price_list:
    tmp = data[data['Price']==price]['Average User Rating'].dropna()
    rating_stats['mean'][price] = tmp.mean()
    rating_stats['std'][price] = tmp.std()
    rating_stats['count'][price] = len(tmp)
```
